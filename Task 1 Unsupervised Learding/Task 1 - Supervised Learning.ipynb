```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

### Importing the file

```python
df = pd.read_csv('http://bit.ly/w-data ')
df
```

```python
plt.scatter(df['Hours'],df['Scores'])
plt.xlabel('Hours')
plt.ylabel('Scores')
```

Splitting the values from target

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

Training the model

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train,y_train)
```

```python
plt.figure()
line = reg.coef_*X + reg.intercept_
plt.scatter(X,y)
plt.plot(X,line)
plt.show()
```

Checking the score

```python
from sklearn.metrics import r2_score,mean_absolute_error
predict = reg.predict(X_test)
print('R2 Score = {}'.format(r2_score(y_test,predict)))
print('Mean Absolute Error = {}'.format(mean_absolute_error(y_test,predict)))
```

Predicting value of 9.25

```python
hours = np.array(9.25)
own_pred = reg.predict(hours.reshape(1,-1))
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
